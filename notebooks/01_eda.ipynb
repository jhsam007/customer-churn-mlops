```python
import sys
from pathlib import Path

PROJECT_ROOT = Path().resolve().parent
sys.path.append(str(PROJECT_ROOT))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
df = pd.read_csv(PROJECT_ROOT / "data/raw/Telco-Customer-Churn.csv")
df.head()
```

```python
df.info()
df.describe(include="all").T
```

```python
#Target Distribution
churn_rate = df["Churn"].value_counts(normalize=True)
churn_rate.plot(kind="bar", title="Churn Distribution")
plt.show()

print(churn_rate)
```

```python
#Key Business Drivers for Churn
sns.boxplot(x="Churn", y="tenure", data=df)
plt.title("Tenure vs Churn")
plt.show()

sns.boxplot(x="Churn", y="MonthlyCharges", data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

```python
#Contract Type
contract_churn = (
    df.groupby("Contract")["Churn"]
    .value_counts(normalize=True)
    .rename("rate")
    .reset_index()
)

sns.barplot(x="Contract", y="rate", hue="Churn", data=contract_churn)
plt.title("Churn Rate by Contract Type")
plt.show()
```

### Key Insights
- Short tenure customers churn more
- Month-to-month contracts have the highest churn
- Higher monthly charges correlate with churn

EDA was used to understand business drivers, not to engineer features.
